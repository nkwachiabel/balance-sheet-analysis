```python
import pandas as pd
```

```python
trial_balance = pd.read_excel("Vinatics 2020 Trial Balance.xlsx")
trial_balance.head()
```

```python
trial_balance['Account type'].unique()
```

```python
assets = trial_balance[trial_balance['Account type'] == 'Assets']
liabilities = trial_balance[trial_balance['Account type'] == 'Liabilities']
equity = trial_balance[trial_balance['Account type'] == 'Equity']
expense = trial_balance[trial_balance['Account type'] == 'Expenses']
revenue = trial_balance[trial_balance['Account type'] == 'Revenue']
```

```python
assets_new = assets.copy()
assets_new = assets_new.drop('Account Code', axis=1)
assets_new = assets_new.groupby(by = ['Account Class']).sum()
assets_new = assets_new.sort_values(by = ['Closing Balance'])
assets_new = pd.DataFrame(assets_new.to_records())
total_assets = {'Account Class': 'Total Assets', 'Opening Balance': assets_new['Opening Balance'].sum(), 'Closing Balance':assets_new['Closing Balance'].sum()}
assets_new = assets_new.append(total_assets, ignore_index = True)
assets_new = assets_new.append(pd.Series(), ignore_index = True)
assets_new
```

```python
liabilities_new = liabilities.copy()
liabilities_new = liabilities_new.drop('Account Code', axis=1)
liabilities_new = liabilities_new.groupby(by = ['Account Class']).sum()
liabilities_new = liabilities_new.sort_values(by = ['Closing Balance'])
liabilities_new = pd.DataFrame(liabilities_new.to_records())
liabilities_new[['Opening Balance','Closing Balance']] *= -1
total_liabilities = {'Account Class': 'Total Liabilities', 'Opening Balance': liabilities_new['Opening Balance'].sum(), 'Closing Balance':liabilities_new['Closing Balance'].sum()}
liabilities_new = liabilities_new.append(total_liabilities, ignore_index = True)
liabilities_new = liabilities_new.append(pd.Series(), ignore_index = True)
liabilities_new
```

```python
equity_new = equity.copy()
equity_new = equity_new.drop('Account Code', axis=1)
equity_new = equity_new.groupby(by = ['Account Class']).sum()
equity_new = equity_new.sort_values(by = ['Closing Balance'])
equity_new = pd.DataFrame(equity_new.to_records())
equity_new[['Opening Balance','Closing Balance']] *= -1
total_equity = {'Account Class': 'Total Equity', 'Opening Balance': equity_new['Opening Balance'].sum(), 'Closing Balance':equity_new['Closing Balance'].sum()}
equity_new = equity_new.append(total_equity, ignore_index = True)
equity_new = equity_new.append(pd.Series(), ignore_index = True)
equity_new
```

```python
expense_new = expense.copy()
expense_new = expense_new.groupby(by = ['Account Class']).sum()
expense_new = pd.DataFrame(expense_new.to_records())
expense_new[['Opening Balance','Closing Balance']] *= -1
total_expense = {'Account Class': 'Total Expense', 'Opening Balance': expense_new['Opening Balance'].sum(), 'Closing Balance':expense_new['Closing Balance'].sum()}
expense_new = expense_new.append(total_expense, ignore_index = True)
expense_new
```

```python
revenue_new = revenue.copy()
revenue_new = revenue_new.groupby(by = ['Account Class']).sum()
revenue_new = pd.DataFrame(revenue_new.to_records())
revenue_new[['Opening Balance','Closing Balance']] *= -1
total_revenue = {'Account Class': 'Total Revenue', 'Opening Balance': revenue_new['Opening Balance'].sum(), 'Closing Balance':revenue_new['Closing Balance'].sum()}
revenue_new = revenue_new.append(total_revenue, ignore_index = True)
revenue_new
```

```python
net_income = revenue_new.iloc[[-1]].append(expense_new.iloc[[-1]], ignore_index=True)
net_income_row = {'Account Class': 'Net Income', 'Opening Balance': net_income['Opening Balance'].sum(), 'Closing Balance':net_income['Closing Balance'].sum()}
net_income = net_income.append(net_income_row, ignore_index=True)
net_income
```

```python
total_liab_and_equity = liabilities_new.iloc[[-2]].append(equity_new.iloc[[-2]], ignore_index=True)
total_liab_and_equity_row = {'Account Class': 'Total Liability and Equity', 'Opening Balance': total_liab_and_equity['Opening Balance'].sum(), 'Closing Balance':total_liab_and_equity['Closing Balance'].sum()}
total_liab_and_equity = total_liab_and_equity.append(total_liab_and_equity_row, ignore_index=True)
total_liab_and_equity
```

```python
total_liab_equity_and_income = total_liab_and_equity.iloc[[-1]].append(net_income.iloc[[-1]], ignore_index=True, sort=False)
total_liab_equity_and_income_row = {'Account Class': 'Total Liability, Equity and Income', 'Opening Balance': total_liab_equity_and_income['Opening Balance'].sum(), 'Closing Balance':total_liab_equity_and_income['Closing Balance'].sum()}
total_liab_equity_and_income = total_liab_equity_and_income.append(total_liab_equity_and_income_row, ignore_index=True)
total_liab_equity_and_income
```

```python
balance_sheet = assets_new.append([liabilities_new, equity_new, total_liab_equity_and_income], ignore_index=True, sort=False)
balance_sheet['Change Amount'] = balance_sheet['Closing Balance'] - balance_sheet['Opening Balance']
balance_sheet['Change %'] = balance_sheet['Change Amount']/balance_sheet['Opening Balance']
balance_sheet = balance_sheet.drop('Account Code', axis=1)
balance_sheet
```

```python
writer = pd.ExcelWriter('Balance Sheet Analysis.xlsx', engine = 'xlsxwriter')
balance_sheet.to_excel(writer, sheet_name='SOFP', index=False, startrow=3)

workbook = writer.book
balance_sheet_sheet = writer.sheets['SOFP']

#Excel Formats
money_format = workbook.add_format({'num_format': '_(* #,##0.00_);_(* (#,##0.00);_(* "-"??_);_(@_)','align': 'vjustify','valign': 'right'})
percent_format = workbook.add_format({'num_format': '0%','align': 'vjustify','valign': 'right'})
header_format = workbook.add_format({'bold': True,'valign': 'top','font_size':14})

#Format the income statement sheet
balance_sheet_sheet.set_zoom(110)
balance_sheet_sheet.write(0, 0, 'VINATICS', header_format)
balance_sheet_sheet.write(1, 0, 'SOFP ANALYSIS', header_format)
balance_sheet_sheet.set_column('A:A', 55)
balance_sheet_sheet.set_column('B:D', 30, money_format)
balance_sheet_sheet.set_column('E:E', 15, percent_format)

writer.save()

import os
os.system('start "excel.exe" "Balance Sheet Analysis.xlsx"')
```
